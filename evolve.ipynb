Generate random fenotype with selected configure

```
import yaml
import numpy as np

def generate(path,num):
  with open(path, 'r') as stream:
    try:
       dic = yaml.safe_load(stream)
    except yaml.YAMLError as exc:
        print(exc)
  fen = []
  for _ in range(num):
    
    gen = []      
    for key, value in dic.items():
      if value['type'] == "int":
        gen.append(np.random.randint(value['min'],value['max']))
      else:
        gen.append(np.random.uniform(value['min'],value['max']))
    fen.append(gen)  
  return fen
tmp = generate("/content/kafka.yaml",5)


  
  
```
